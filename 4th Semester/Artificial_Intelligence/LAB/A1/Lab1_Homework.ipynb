```python
# Tensor Operations 

import torch 

# Define two tensors
a = torch.tensor([1, 2, 3]) 
b = torch.tensor([4, 5, 6]) 

# Element-wise addition and multiplication 
print("\nAddition:", a + b) 
print("\nMultiplication:", a * b) 

# Reshaping tensors 
tensor3 = torch.arange(1, 10).reshape(3, 3 ) 
print("\nReshaped Tensor:\n", tensor3) 

#Accessing elements 
print("\nElement at row 1, col 2:" , tensor3[1,2]) 
print("\nFirst column:\n", tensor3[:, 0]) 


# Convert Between Numpy & Pytorch 
np_array = tensor3.numpy() 
print("\nConverted back to Numpy:\n", np_array)

# Convert Numpy array back to Pytorch tensor
tensor_from_np = torch.from_numpy(np_array)
print("\nConverted back to Tensor:\n", tensor_from_np)
```

```python
# Application1: Simulate an unfair probability Spinner 

import numpy as np 
import matplotlib.pyplot as plt 

# Define the outcomes and probabilities 
outcomes = ["Red", "Blue", "Green", "Yellow"] 
probabilities = [0.5, 0.2, 0.2, 0.1] # Unequal probabilities 

# Simulate a spin 
def spin_wheel():
    return np.random.choice(outcomes, p=probabilities)

# Run multiple spins
spins = [spin_wheel() for _ in range(1000)]

# Count occurrences 
from collections import Counter 
result_counts = Counter(spins) 

# Display results
print("Spinner Results:" , result_counts)

# Plot the results
plt.bar(result_counts.keys(), result_counts.values(), color=["red", "blue", "green", "yellow"])
plt.xlabel("Color")
plt.ylabel("Frequency") 
plt.title("Unfair Probability Spinner Results")
plt.show() 
```

```python
# Estimating Pi Using Random points
```

```python
def monte_carlo_pi(n_points=10000):
    inside_circle = 0

    for _ in range(n_points):
        x, y = np.random.uniform(-1, 1, 2) #Generate random (x,y) between -1 and 1 
        if x**2 + y**2 <= 1: 
            inside_circle += 1

    estimated_pi = (inside_circle / n_points) * 4
    return estimated_pi 


# Run the Monte Carlo estimation of Pi 
estimated_pi = monte_carlo_pi(10000)
print("Estimated Pi Value:", estimated_pi)

    

#Visualizing Monte Carlo Pi approximation 
n_points = 5000
inside_x, inside_y, outside_x, outside_y = [], [], [], []

for _ in range(n_points):
    x, y = np.random.uniform(-1, 1, 2) 
    if x**2 + y**2 <= 1:
        inside_x.append(x)
        inside_y.append(y) 
    else: 
        outside_x.append(x)
        outside_y.append(y)

plt.figure(figsize=(6,6))
plt.scatter(inside_x, inside_y, color="blue", s=1, label="Inside Circle") 
plt.scatter(outside_x, outside_y, color="red", s=1, label="Outside Circle") 
plt.legend() 
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Monte Carlo Simulation of Pi") 


    


```
